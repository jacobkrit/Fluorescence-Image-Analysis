```python
def Read_and_Calculate_Distribution_Features(slice_dir,slice_numbers):
    vals_list = []
    skew_list = []
    kurt_list = []
    for slice_step in range(1,slice_numbers+1):      
        distribution_saved_path = "D:/Ultrasound/Slices-Distributions/"+slice_dir+"-"+str(slice_step)+".mat"
        slice_trackdata = io.loadmat(distribution_saved_path)
        slice_data = slice_trackdata['saved_data'][0];
        vals_list.append(slice_data)
        skew_list.append(skew(slice_data))
        kurt_list.append(kurtosis(slice_data))
    return vals_list, skew_list, kurt_list

#vals_, skew_, kurt_ = Read_and_Calculate_Distribution_Features("",32)
```

A standard normal distribution has kurtosis of 3 and is recognized as mesokurtic. An increased kurtosis (>3) can be 
visualized as a thin bell with a high peak whereas a decreased kurtosis corresponds to a broadening of the peak and thickening of the tails. Kurtosis >3 is recognized as leptokurtic and <3 as platykurtic
